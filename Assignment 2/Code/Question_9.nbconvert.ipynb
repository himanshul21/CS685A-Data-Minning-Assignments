```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#Loading the dataset and preprocessing
df = pd.read_excel("../Dataset/Input_Dataset/DDW-C19-0000.xlsx", header = [1])
df = df[df.iloc[:, 3] == "Total"]
df = df.iloc[:, [0, 4, 6, 7, 9, 10]]

#Droping the row which are not important for analysis
df = df[df["Educational level"] != "Total"]
df = df[df["Educational level"] != "Literate"]

#Renaming the column names
df.columns = ["State", "Educational Level", "male two lang", "female two lang", "male three lang", "female three lang"]
df.head(6)
```

```python
#Loading the dataset and preprocessing
df_08= pd.read_excel("../Dataset/Input_Dataset/DDW-0000C-08.xlsx", header=[2])

#Column filtering
df_08 = df_08.iloc[4:, [1, 4, 5, 10, 11, 19, 20, 22, 23, 25, 26, 28, 29, 31, 32, 34, 35, 37, 38, 40, 41]]

#Renaming the column name
df_08.columns = ["State", "Total", "Age-group" , "Male Illiterate", "Female Illiterate", "Male Literate but below primary",
                 "Female Literate but below primary", "Male Primary but below middle", "Female Primary but below middle", 
                 "Male Middle but below matric/secondary", "Female Middle but below matric/secondary", "Male Matric/Secondary", 
                 "Female Matric/Secondary", "Male Higher Secondary", "Female Higher Secondary", "Male Non Technical Diploma", 
                 "Female Non Technical Diploma", "Male Technical Diploma", "Female Technical Diploma", "Male Graduate and above", 
                 "Female Graduate and above"]      

#Calculating the population of male nad female Matric/Secondary but below graduate
df_08["Male Matric/Secondary but below graduate"] = df_08["Male Matric/Secondary"] + df_08["Male Higher Secondary"] + df_08["Male Non Technical Diploma"] + df_08["Male Technical Diploma"]
df_08["Female Matric/Secondary but below graduate"] = df_08["Female Matric/Secondary"] + df_08["Female Higher Secondary"] + df_08["Female Non Technical Diploma"] + df_08["Female Technical Diploma"]

#Droping unimportant rows 
df_08 = df_08[df_08["Total"] == "Total"]
df_08 = df_08[df_08["Age-group"] == "All ages"]
df_08.drop(["Total", "Age-group", "Male Matric/Secondary", "Male Higher Secondary", "Male Non Technical Diploma", "Male Technical Diploma", 
           "Female Matric/Secondary", "Female Higher Secondary", "Female Non Technical Diploma", "Female Technical Diploma"], axis=1, inplace=True)

#Reordering the index
df_08.index = list(range(len(df_08)))
df_08.head()
```

```python
category = list(df["Educational Level"].unique())
```

```python
unique_st = list(df["State"].unique())
```

```python
dict = {"State" : [],
       "Educational Level" : [],
       "Total Male" : [],
       "Total Female" : []}
```

```python
#Creating new dataframe for education level for male and female
for st in unique_st:
    for cat in category:
        col_male = "Male " + str(cat)
        col_female = "Female " + str(cat)
        
        male_count = df_08[(df_08["State"] == int(st)) & (df_08[col_male])][col_male].iloc[0]
        female_count = df_08[(df_08["State"] == int(st)) & (df_08[col_female])][col_female].iloc[0]
        
        dict["State"].append(st)
        dict["Educational Level"].append(cat)
        dict["Total Male"].append(male_count)
        dict["Total Female"].append(female_count)
```

```python
df_08_new = pd.DataFrame(dict)
```

```python
df_08_new[df_08_new["State"] == "00"]
```

```python
df_08_new[df_08_new["State"] == "01"]
```

```python
df_final = pd.merge(df, df_08_new, on = ["State", "Educational Level"])
```

```python
df_final
```

```python
#Calculating the male and female population for speaking exactly one, two and three lang
df_final["Male exact one language"] = df_final["Total Male"] - df_final["male two lang"]
df_final["Female exact one language"] = df_final["Total Female"] - df_final["female two lang"]
df_final["Male exact two language"] = df_final["male two lang"] - df_final["male three lang"]
df_final["Female exact two language"] = df_final["female two lang"] - df_final["female three lang"]
df_final = df_final.rename(columns={"male three lang" : "Male exact three language", "female three lang" : "Female exact three language"})
df_final.drop(["male two lang", "female two lang"], axis=1, inplace=True)
```

```python
df_final
```

```python
#Calculating the male and female ratio for speaking exactly one, two and three lang
df_final["Male 1 Ratio"] = df_final["Male exact one language"] / df_final["Total Male"]
df_final["Female 1 Ratio"] = df_final["Female exact one language"] / df_final["Total Female"]

df_final["Male 2 Ratio"] = df_final["Male exact two language"] / df_final["Total Male"]
df_final["Female 2 Ratio"] = df_final["Female exact two language"] / df_final["Total Female"]

df_final["Male 3 Ratio"] = df_final["Male exact three language"] / df_final["Total Male"]
df_final["Female 3 Ratio"] = df_final["Female exact three language"] / df_final["Total Female"]
```

```python
gb_final = df_final.groupby(by = "State")
```

```python
#Calculating the literacy group for male speaking exactly one lang
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[0]:, "Male 1 Ratio"]) + value[0])
index.sort()
df_1_male = df_final.loc[index, ["State", "Educational Level", "Male 1 Ratio"]]
df_1_male.head()
```

```python
#Calculating the literacy group for female speaking exactly one lang
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[0]:, "Female 1 Ratio"]) + value[0])
index.sort()
df_1_female = df_final.loc[index, ["State", "Educational Level", "Female 1 Ratio"]]
df_1_female.head()
```

```python
#Calculating the literacy group for male speaking exactly two lang
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[0]:, "Male 2 Ratio"]) + value[0])
index.sort()
df_2_male = df_final.loc[index, ["State", "Educational Level", "Male 2 Ratio"]]
df_2_male.head()
```

```python
#Calculating the literacy group for female speaking exactly two lang
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[0]:, "Female 2 Ratio"]) + value[0])
index.sort()
df_2_female = df_final.loc[index, ["State", "Educational Level", "Female 2 Ratio"]]
df_2_female.head()
```

```python
#Calculating the literacy group for male speaking exactly three lang
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[0]:, "Male 3 Ratio"]) + value[0])
index.sort()
df_3_male = df_final.loc[index, ["State", "Educational Level", "Male 3 Ratio"]]
df_3_male.head()
```

```python
#Calculating the literacy group for female speaking exactly three lang
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[0]:, "Female 3 Ratio"]) + value[0])
index.sort()
df_3_female = df_final.loc[index, ["State", "Educational Level", "Female 3 Ratio"]]
df_3_female.head()
```

```python
#Creating the dataframe for speaking exactly one lang
literacy_gender_1 = pd.merge(df_1_male, df_1_female, on = ["State"])
literacy_gender_1.columns = ["state/ut", "literacy-group-males", "ratio-males", "literacy-group-females", "ratio-females"]
literacy_gender_1.head()
```

```python
#Creating the dataframe for speaking exactly two lang
literacy_gender_2 = pd.merge(df_2_male, df_2_female, on = ["State"])
literacy_gender_2.columns = ["state/ut", "literacy-group-males", "ratio-males", "literacy-group-females", "ratio-females"]
literacy_gender_2.head()
```

```python
#Creating the dataframe for speaking exactly three lang
literacy_gender_3 = pd.merge(df_3_male, df_3_female, on = ["State"])
literacy_gender_3.columns = ["state/ut", "literacy-group-males", "ratio-males", "literacy-group-females", "ratio-females"]
literacy_gender_3.head()
```

```python
#Storing the final result
literacy_gender_1.to_csv("../Dataset/Output_Dataset/literacy-gender-c.csv", index = False)
literacy_gender_2.to_csv("../Dataset/Output_Dataset/literacy-gender-b.csv", index = False)
literacy_gender_3.to_csv("../Dataset/Output_Dataset/literacy-gender-a.csv", index = False)
```
