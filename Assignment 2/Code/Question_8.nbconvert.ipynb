```python
import numpy as np
import pandas as pd
```

```python
#Loading the dataset and preprocessing
df = pd.read_excel("../Dataset/Input_Dataset/DDW-C18-0000.xlsx", header = [1])
df = df[df.iloc[:, 3] == "Total"]
df = df.iloc[:, [0, 2, 4, 6, 7, 9, 10]]
#Droping the unimportant columns
df = df[df["Age-group"] != "Total"]
df = df[df["Age-group"] != "Age not stated"]
#Renaming the column name
df.columns = ["State Code", "Name", "Age-group", "Male two lang", "Female two lang", "Male three lang", "Female three lang"]
df.head(8)
```

```python
#Loading the dataset and preprocessing
df_14 = pd.read_excel("../Dataset/Input_Dataset/DDW-0000C-14.xls", header=[1])
#Renaming the column name
df_14 = df_14.rename(columns={"Area Name": "Name", " State Code" : "State Code"})
df_14 = df_14.iloc[5:, [1, 3, 4, 6, 7]]
df_14 = df_14.rename(columns={"Unnamed: 6" : "Total Male", "Unnamed: 7" : "Total Female"})
#Droping the unimportant rows
df_14 = df_14[df_14["Age-group"] != "All ages"]
df_14 = df_14[df_14["Age-group"] != "Age not stated"]
#Transforming the state name
df_14.loc[24:, "Name"] = df_14.loc[24:, "Name"].apply(lambda x: x.split("-")[1].split("(")[0].strip()) 
df_14.replace("India", "INDIA", inplace=True)
#Reordering the index
df_14.index = list(range(len(df_14.index)))
df_14.head(17)
```

```python
df_14_final = df_14.copy()
```

```python
#Making groupby using Name column
gb14 = df_14_final.groupby("Name")
```

```python
#Calculating the population of male and female for unknown age-group 
for key, value in gb14.groups.items():
    age30_49M = np.sum(df_14_final.loc[gb14.groups[key][6:10], "Total Male"])
    age50_69M = np.sum(df_14_final.loc[gb14.groups[key][10:14], "Total Male"])
    age70_plusM = np.sum(df_14_final.loc[gb14.groups[key][14:17], "Total Male"])
    
    age30_49f = np.sum(df_14_final.loc[gb14.groups[key][6:10], "Total Female"])
    age50_69f = np.sum(df_14_final.loc[gb14.groups[key][10:14], "Total Female"])
    age70_plusf = np.sum(df_14_final.loc[gb14.groups[key][14:17], "Total Female"])
    
    st_code = df_14_final[df_14_final["Name"] == key]["State Code"].iloc[0]
    df2= {"State Code" : st_code, "Name" : key, "Age-group" : "30-49", "Total Male" : age30_49M, "Total Female" :age30_49f}
    df3= {"State Code" : st_code, "Name" : key, "Age-group" : "50-69", "Total Male" :age50_69M, "Total Female" :age50_69f}
    df4= {"State Code" : st_code, "Name" : key, "Age-group" : "70+", "Total Male": age70_plusM, "Total Female" :age70_plusf}
    df_14_final = df_14_final.append(df2, ignore_index=True)
    df_14_final = df_14_final.append(df3, ignore_index=True)
    df_14_final = df_14_final.append(df4, ignore_index=True)
```

```python
#Droping the unimportant rows
df_14_final = df_14_final[df_14_final["Age-group"] != "0-4"]
df_14_final = df_14_final[df_14_final["Age-group"] != "30-34"]
df_14_final = df_14_final[df_14_final["Age-group"] != "35-39"]
df_14_final = df_14_final[df_14_final["Age-group"] != "40-44"]
df_14_final = df_14_final[df_14_final["Age-group"] != "45-49"]
df_14_final = df_14_final[df_14_final["Age-group"] != "50-54"]
df_14_final = df_14_final[df_14_final["Age-group"] != "55-59"]
df_14_final = df_14_final[df_14_final["Age-group"] != "60-64"]
df_14_final = df_14_final[df_14_final["Age-group"] != "65-69"]
df_14_final = df_14_final[df_14_final["Age-group"] != "70-74"]
df_14_final = df_14_final[df_14_final["Age-group"] != "75-79"]
df_14_final = df_14_final[df_14_final["Age-group"] != "80+"]
```

```python
df_14_final[df_14_final["Name"] == "INDIA"]
```

```python
#Loading the dataset and preprocessing
```

```python
df_final = pd.merge(df, df_14_final, on = ["State Code", "Name", "Age-group"])
df_final["Male exact one language"] = df_final["Total Male"] - df_final["Male two lang"]
df_final["Female exact one language"] = df_final["Total Female"] - df_final["Female two lang"]
df_final["Male exact two language"] = df_final["Male two lang"] - df_final["Male three lang"]
df_final["Female exact two language"] = df_final["Female two lang"] - df_final["Female three lang"]
df_final = df_final.rename(columns={"Male three lang" : "Male exact three language", "Female three lang" :"Female exact three language"})
df_final.drop(["Male two lang", "Female two lang"], axis=1, inplace = True)
df_final.head(10)
```

```python
#Calculating the male and female ratio for exactly speaking one , two and three language

df_final["Male 1 Ratio"] = df_final["Male exact one language"] / df_final["Total Male"]
df_final["Female 1 Ratio"] = df_final["Female exact one language"] / df_final["Total Female"]

df_final["Male 2 Ratio"] = df_final["Male exact two language"] / df_final["Total Male"]
df_final["Female 2 Ratio"] = df_final["Female exact two language"] / df_final["Total Female"]

df_final["Male 3 Ratio"] = df_final["Male exact three language"] / df_final["Total Male"]
df_final["Female 3 Ratio"] = df_final["Female exact three language"] / df_final["Total Female"]
```

```python
df_final[df_final["State Code"] == "01"]
```

```python
gb_final = df_final.groupby(by = "Name")
```

```python
#Calculating the male age-group speaking one lang
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[0]:, "Male 1 Ratio"]) + value[0])
index.sort()
df_1_male = df_final.loc[index, ["State Code", "Age-group", "Male 1 Ratio"]]
df_1_male.head()
```

```python
#Calculating the female age-group speaking one lang
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[0]:, "Female 1 Ratio"]) + value[0])
index.sort()
df_1_female = df_final.loc[index, ["State Code", "Age-group", "Female 1 Ratio"]]
df_1_female.head()
```

```python
#Calculating the male age-group speaking two lang
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[0]:, "Male 2 Ratio"]) + value[0])
index.sort()
df_2_male = df_final.loc[index, ["State Code", "Age-group", "Male 2 Ratio"]]
df_2_male.head()
```

```python
#Calculating the female age-group speaking two lang
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[0]:, "Female 2 Ratio"]) + value[0])
index.sort()
df_2_female = df_final.loc[index, ["State Code", "Age-group", "Female 2 Ratio"]]
df_2_female.head()
```

```python
#Calculating the male age-group speaking three lang
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[0]:, "Male 3 Ratio"]) + value[0])
index.sort()
df_3_male = df_final.loc[index, ["State Code", "Age-group", "Male 3 Ratio"]]
df_3_male.head()
```

```python
#Calculating the female age-group speaking three lang
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[0]:, "Female 3 Ratio"]) + value[0])
index.sort()
df_3_female = df_final.loc[index, ["State Code", "Age-group", "Female 3 Ratio"]]
df_3_female.head()
```

```python
#Creating the dataframe for speaking exactly one lang
age_gender_1 = pd.merge(df_1_male, df_1_female, on = ["State Code"])
#Renamin the column name
age_gender_1.columns = ["state/ut", "age-group-males", "ratio-males", "age-group-females", "ratio-females"]
age_gender_1.head()
```

```python
#Creating the dataframe for speaking exactly two lang
age_gender_2 = pd.merge(df_2_male, df_2_female, on = ["State Code"])
#Renamin the column name
age_gender_2.columns = ["state/ut", "age-group-males", "ratio-males", "age-group-females", "ratio-females"]
age_gender_2.head()
```

```python
#Creating the dataframe for speaking exactly three lang
age_gender_3 = pd.merge(df_3_male, df_3_female, on = ["State Code"])
#Renamin the column name
age_gender_3.columns = ["state/ut", "age-group-males", "ratio-males", "age-group-females", "ratio-females"]
age_gender_3.head()
```

```python
#Storing the final result
age_gender_1.to_csv("../Dataset/Output_Dataset/age-gender-c.csv", index = False)
age_gender_2.to_csv("../Dataset/Output_Dataset/age-gender-b.csv", index = False)
age_gender_3.to_csv("../Dataset/Output_Dataset/age-gender-a.csv", index = False)
```
