```python
#Loading the dataset and preprocessing
df_14 = pd.read_excel("../Dataset/Input_Dataset/DDW-0000C-14.xls", header=[1])
df_14 = df_14.rename(columns={"Area Name": "Name"})
df_14 = df_14.iloc[5:, [1, 3, 4, 5]]

#Neglacting the rows with all ages and age not stated
df_14 = df_14[df_14["Age-group"] != "All ages"]
df_14 = df_14[df_14["Age-group"] != "Age not stated"]

#filtering the state name
df_14.loc[24:, "Name"] = df_14.loc[24:, "Name"].apply(lambda x: x.split("-")[1].split("(")[0].strip()) 

#Reordering the index
df_14.index = list(range(len(df_14.index)))
```

```python
#Making the copy of original dataframe
df_14_final = df_14.copy()
```

```python
#Making groupby using name
gb14 = df_14_final.groupby("Name")
```

```python
#Calculaitng the rows for age 30-49, 50-69, 70+ and apending it
for key, value in gb14.groups.items():
    age30_49 = np.sum(df_14_final.loc[gb14.groups[key][6:10], "Total "])
    age50_69 = np.sum(df_14_final.loc[gb14.groups[key][10:14], "Total "])
    age70_plus = np.sum(df_14_final.loc[gb14.groups[key][14:17], "Total "])
    st_code = df_14_final[df_14_final["Name"] == key][" State Code"].iloc[0]
    df2= {" State Code" : st_code, "Name" : key, "Age-group" : "30-49", "Total " : age30_49}
    df3= {" State Code" : st_code, "Name" : key, "Age-group" : "50-69", "Total " : age50_69}
    df4= {" State Code" : st_code, "Name" : key, "Age-group" : "70+", "Total " : age70_plus}
    df_14_final = df_14_final.append(df2, ignore_index=True)
    df_14_final = df_14_final.append(df3, ignore_index=True)
    df_14_final = df_14_final.append(df4, ignore_index=True)
```

```python
#Droping unimportnat columns
df_14_final = df_14_final[df_14_final["Age-group"] != "0-4"]
df_14_final = df_14_final[df_14_final["Age-group"] != "30-34"]
df_14_final = df_14_final[df_14_final["Age-group"] != "35-39"]
df_14_final = df_14_final[df_14_final["Age-group"] != "40-44"]
df_14_final = df_14_final[df_14_final["Age-group"] != "45-49"]
df_14_final = df_14_final[df_14_final["Age-group"] != "50-54"]
df_14_final = df_14_final[df_14_final["Age-group"] != "55-59"]
df_14_final = df_14_final[df_14_final["Age-group"] != "60-64"]
df_14_final = df_14_final[df_14_final["Age-group"] != "65-69"]
df_14_final = df_14_final[df_14_final["Age-group"] != "70-74"]
df_14_final = df_14_final[df_14_final["Age-group"] != "75-79"]
df_14_final = df_14_final[df_14_final["Age-group"] != "80+"]
```

```python
df_14_final[df_14_final["Name"] == "India"]
```

```python
#Finding the unique state
unique_st_code = list(df_14_final[" State Code"].unique())
```

```python
#Finding the unique age group
unique_age_group = list(df_final["Age-group"].unique())
```

```python
#Creating new dictionary
dict = {"state/ut" : [],
       "age-group" : [],
       "percentage" : []}
```

```python
#Calculating the percentage for each group
for st_code in unique_st_code:
    for age_group in unique_age_group:
        dict["state/ut"].append(st_code)
        dict["age-group"].append(age_group)
        count = int(df_final[(df_final["State"] == st_code) & (df_final["Age-group"] == age_group)]["Number speaking third language"].iloc[0])
        total = int(df_14_final[(df_14_final[" State Code"] == st_code) & (df_14_final["Age-group"] == age_group)]["Total "].iloc[0])
        percentage = count * 100 / total
        dict["percentage"].append(percentage)
```

```python
df_ = pd.DataFrame(dict)
```

```python
df_.head(8)
```

```python
gb_final = df_.groupby(by = "state/ut")
```

```python
#Finding the maximum percentage age-group
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[1]:, "percentage"]) + value[1])
index.sort()
df_new = df_.iloc[index]
df_new
```

```python
#Storing the final result
df_new.to_csv("../Dataset/Output_Dataset/age-india.csv", index = False)
```
