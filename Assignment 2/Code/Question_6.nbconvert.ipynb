```python
#Importing the libraries
import numpy as np
import pandas as pd
```

```python
#Importing the dataset and preprocessing
df = pd.read_excel("../Dataset/Input_Dataset/DDW-C19-0000.xlsx", header = [1])
df = df.rename(columns={"Area Name" : "Name"})
df = df[df.iloc[:, 3] == "Total"]
df = df.iloc[:, [0, 4, 8]]

#Droping the unimportant rows
df = df[df["Educational level"] != "Total"]
df = df[df["Educational level"] != "Literate"]
df.head(6)
```

```python
df_final = df.copy()
df_final.index = np.arange(0, len(df))
```

```python
#importing the dataset and preprocessing
df_08= pd.read_excel("../Dataset/Input_Dataset/DDW-0000C-08.xlsx", header=[2])
df_08 = df_08.iloc[4:, [1, 4, 5, 9, 18, 21, 24, 27, 30, 33, 36, 39]]

#Renaming the columns
df_08.columns = ["State Code", "Total", "Age-group" , "Illiterate", "Literate but below primary", "Primary but below middle",
                 "Middle but below matric/secondary", "Matric/Secondary", "Higher Secondary", "Non Technical Diploma", 
                 "Technical Diploma", "Graduate and above"]
df_08["Matric/Secondary but below graduate"] = df_08["Matric/Secondary"] + df_08["Higher Secondary"] + df_08["Non Technical Diploma"] + df_08["Technical Diploma"]

#Filtering the rows
df_08 = df_08[df_08["Total"] == "Total"]
df_08 = df_08[df_08["Age-group"] == "All ages"]
#Droping the umimportant columns
df_08.drop(["Total", "Age-group", "Matric/Secondary", "Higher Secondary", "Non Technical Diploma", "Technical Diploma"], axis=1, inplace=True)
#Reordering the index
df_08.index = list(range(len(df_08)))
df_08
```

```python
df_08_final = df_08.copy()
```

```python
#Finding the unique state
unique_st_code = df_final["State"].unique()
unique_st_code
```

```python
#Finding the unique literacy group
unique_literacy = list(df_final["Educational level"].unique())
```

```python
#Creating new dictonary
dict = {"state/ut" : [], 
       "literacy-group" : [], 
       "percentage" : []}
```

```python
#Calculating the percentage for each literacy group
for st_code in unique_st_code:
    for category in unique_literacy:
        count = int(df_final[(df_final["State"] == st_code) & (df_final["Educational level"] == category)]["Number speaking third language"].iloc[0])
        total = int(df_08_final.loc[int(st_code), category])
        percentage = count * 100 / total
        dict["state/ut"].append(st_code)
        dict["literacy-group"].append(category)
        dict["percentage"].append(percentage)
```

```python
final_df = pd.DataFrame(dict)
```

```python
#Making groupby usning state code
gb_final = final_df.groupby("state/ut")
```

```python
#Finding the maximum percentage literacy-group
index = []
for key, value in gb_final.groups.items():
    index.append(np.argmax(gb_final.get_group(key).loc[value[1]:, "percentage"]) + value[1])
index.sort()
df_new = final_df.iloc[index]
df_new
```

```python
#Storing the final result
df_new.to_csv("../Dataset/Output_Dataset/literacy-india.csv", index = False)
```
